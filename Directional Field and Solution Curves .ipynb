Considering an Initial Value Problem (IVP) of the ODE  function $y' = f(x,y)$ with initial condition at $x = 0$ written as 

$$y' = f(x,y) \quad y(0) = y_{0} $$ $ \\ $Assuming our function is continous in the interval $I$ and function $f : I \longrightarrow \mathbb{R}$. $\\$We can solve the following Ordinary Diffrential Equation (ODE) by Integrating the function$f(y,x)$ through out the interval $I$ from $ t_{0}$ to $t$ which is given as.
$$ \Psi = \int_{0}^x g(s)ds $$  After Integrating we get the solution as $$ y(t) = \Psi +C $$ where C $\in \mathbb{R}$ and this equation is also called as Genral Solution of an ODE different values of C gives different solutions all these solutions can be seen in directional field. If we substitute our initial condition at $x = 0 , y = y_{0}$ we get a unique value of C which can be denoted as $\delta$ then,

$$y(t) = \Psi + \delta$$
Let us take the ODE with $$ y' = x \quad y(0) = 0 $$ for this the function $f(y,x) = x$ where the function $f$ is continious in $\mathbb{R}$ for the genral solution of this differential equation we can integrate our function through out the interval $I$ where $I \in [\ 0 , x ]\ $ we get,$$ y(x) = \int_{0}^x s ds  $$ the Genral Solution as   $$ y(x) = \frac{x^2}{2} + C $$ Now from the IVP $y(0) = 0$ we get, $$y(t) = \frac{x^2}{2}$$
